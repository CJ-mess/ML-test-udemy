```python
# przygotowanie danych do modelu

import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler

sklearn.__version__

data =  {'size': ['XL', 'L', 'M', 'L', 'M'],
         'color': ['RED', 'GREEN', 'BLUE', 'GREEN', 'RED'],
         'gender': ['FEMALE', 'MALE', 'MALE', 'FEMALE', 'FEMALE'],
         'price': [199.0, 89.0, 99.0, 129.0, 79.0],
         'weight':[500, 450, 300, 380, 410],
         'bought': ['yes', 'no', 'yes', 'no','yes']}

df_raw = pd.DataFrame(data=data)
df_raw
df = df_raw.copy()
df.info()

for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

# utworzenie klasy label encoder, ktora pozwoli na zakodowanie zmiennej docelowej 'bought'
# przygotowanie danych do modelu rozpoczynamy wiec od konca, od przygotowania danej wynikowej
# wynikiem konwersji powinno byc 0 (nie da sie / nie spelnia wyniku) lub 1 (da sie / spelnia wynik)
le = LabelEncoder()
df['bought'] = le.fit_transform(df['bought'])

# nastpnie uzywamy scalera sla zmiennych numerycznych. Tak, zeby bylo one w formie odchylen standardowych
# najlepiej wszystkie na raz
# wynik konwersji to odchylenia standardowe
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

# na koniec zostaja dane tekstowe (kategoryczne), ktore skonwertuja je na numeryczne
# uzywamy 'drop_first' w celu usuniecia pierwszych kolumn dla 'get_dummies'. Tak, zeby nie powodowala dalszego konfliktu danych
# wynik konwersji to 0 lub 1 dla wybranej nowo stworzonej kategorii w kolumnach
df = pd.get_dummies(data=df, drop_first=True)
df
```
